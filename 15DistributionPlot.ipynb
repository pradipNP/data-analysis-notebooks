```python
import seaborn as sns
import numpy as np

np.random.seed(0)
x = np.random.randn(100)
ax_1 = sns.distplot(x)
```

```python
ax_2 = sns.distplot(x, rug = True, hist = False)
```

```python
x_2 = sns.distplot(x, rug = False, hist = True)
```

```python
from scipy.stats import norm
ax_3 = sns.distplot(x, fit = norm)
```

```python
ax_3 = sns.distplot(x, fit = norm, kde = False) # (kde=kernel density estimatation)
```

```python
# Changing to verical

ax_4 = sns.distplot(x, vertical = True)
```

```python
# Changing color
sns.set_color_codes()
ax_5 = sns.distplot(x, color = 'r')
```

```python
# CHange background color

sns.set(style = 'dark')
ax_6 = sns.distplot(x, color = 'green')
```
