# MEASURES FOR IN-SAMPLE EVALUATION
- TWO PARAMETERS
- 1. MSE(Mean Square Error)
  2. R-Squared(R^2)

## 1.MSE:
- Average of estimated and actual values

 ## Steps To Follow:
 1. Subtracting: (y = y1 - y^​i)
 2. Squaring: (y^2)
 3. Mean/Average = (adding all of them)/no.of samples

```python
from sklearn.metrics import mean_squared_error
import seaborn as sns
flights = sns.load_dataset("flights")
```

```python
mean_squared_error(flights['year'], flights['passengers'])
```

```python
mean_squared_error(flights['year'], flights['passengers'], multioutput = 'raw_values') # raw_values returns the full set of errors
```

```python
mean_squared_error(flights['year'], flights['passengers'], multioutput = 'uniform_average')
```

```python
mean_squared_error(flights['year'], flights['passengers'], squared = True) 
```

```python
mean_squared_error(flights['year'], flights['passengers'], squared = False) 
# It don't returns us MSE but it returns RMSE
```

## 2. R-squared:
- It determines how close the data is fitted to the regression line
- R^2 = 1 - (MSE of Regression Line)/(MSE of Average of Data)
- Ranges 0 to 1

```python
x = flights[['year']]
y = flights[['passengers']]
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
```

```python
lm.fit(x,y)
```

```python
lm.score(x,y)
```
