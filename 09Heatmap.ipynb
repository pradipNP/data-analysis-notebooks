```python
import numpy as np
np.random.seed(0)
import seaborn as sns #seaborn is used for the heatmap

uniform_data = np.random.rand(5,6) #where 5 is row and 6 is column
print(uniform_data)
```

```python
# heat map

heat_map = sns.heatmap(uniform_data)
```

```python
heat_map_2 = sns.heatmap(uniform_data, vmin = 0, vmax = 1)
```

```python
normal_data = np.random.randn(6,6)
print(normal_data)
```

```python
heat_map_3 = sns.heatmap(normal_data, center = 0)
```

```python
# How to load dataset and modify certain things

flights = sns.load_dataset("flights")
flights = flights.pivot(index="month",columns="year",values="passengers")
heat_map_4 = sns.heatmap(flights)
```

```python
#bringing numbers into the heat map
heat_map_5 = sns.heatmap(flights, annot= True, fmt = 'd')  #annot=True helps in bringing the numbers inside the heat map, && fmt = d means format is integer type
```

```python
heat_map_5 = sns.heatmap(flights, annot= True, fmt = '.1f')  # for float type
```

```python
# Change color scheme

heat_map_6 = sns.heatmap(flights, cmap="YlGnBu")
```

```python
heat_map_7 = sns.heatmap(flights, cmap = "RdBu") #Red and blue
```

```python
# Change position of bar

heat_map_8 = sns.heatmap(flights, cbar=False)
```
