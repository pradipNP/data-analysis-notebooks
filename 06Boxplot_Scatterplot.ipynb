```python
import pandas as pd
path = "D:/tutedude/Data Analysis/CSV_Files/numeric.csv"
df = pd.read_csv(path)
print(df)
```

# Boxplot
- searbon library is used for boxplot

```python
import seaborn as sns
sns.boxplot(x="Length", y="E", data=df)
```

# Scatterplot
- matplotlib library is used for the scatterplot

```python
import matplotlib.pyplot as plt

x = df["Length"]
y = df["E"]

plt.scatter(x,y)
plt.title("Scatterplot along the x and y axis")

plt.xlabel("Length")
plt.ylabel("Width")
```
