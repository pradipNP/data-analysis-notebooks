# CORRELATION
- Measures to what extent different variables are interdependent.
- Example: Lung cancer -> Smoking

```python
import seaborn as sns
flights = sns.load_dataset("flights")
sns.regplot(x="passengers", y = "year", data = flights)
```

## Two Parameter of Correlation
1. Correlation co-efficient
2. P-Value

### 1. Correlation co-efficient:
- Determines strength of relationship between the variables.
- Range is -1 to +1
- Close to +1 : Large Positive Relationship
- Close to -1 : Large Negaitve Relationship
- Close to 0: No Relationship

### 2. P-Value
- Determines the values and also checks whether result is significant or not
  1. If Value < 0.001 , Result = Strong
  2. Value<0.05 = Moderate
  3. Value<0.1 = Weak
  4. Value>0.1 = No Certainty

```python
# To work with correlation we have to import scipy

from scipy import stats
flights = sns.load_dataset("flights")
print(flights)
```

```python
pearson_correlation = stats.pearsonr(flights["year"], flights["passengers"])
print(pearson_correlation)
```

```python
# It is strong correlation since the correlation coefficients value is in range -1 to +1 and P-value also <0.001
```
