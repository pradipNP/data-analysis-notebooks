# Data Normalization
| Not Normalized |       | Normalized |       |
|----------------|-------|------------|-------|
| Age            | Salary| Age        | Salary|
| 30             | 400000| 0.3        | 1     |
| 40             | 60000 | 0.4        | 0.15  |
| 50             | 300000| 0.5        | 0.75  |
- Original raw values of dataset is in Not Normalized form.
- Normalization scales the data to a common range, usually 0 to 1.

## Methods of Data Normalization

```python
import pandas as pd
path = "D:/tutedude/Data Analysis/CSV_Files/numeric.csv"
df = pd.read_csv(path)
print(df)
```

### 1. Simple Feature Scaling Method

```python
# Normalizing the value of Length column by Simple Feature Scaling Method

df["Length"] = df["Length"]/df["Length"].max()
df
```

### 2. Min-Max Method

```python
df["Length"] = (df["Length"]-df["Length"].min())/(df["Length"].max()-df["Length"].min())
df
```

### 3. Z-Score Method

```python
df["Length"] = (df["Length"]-df["Length"].mean())/df["Length"].std()
df
```
