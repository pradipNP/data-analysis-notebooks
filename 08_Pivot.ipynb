## Pivot:
- Pivot helps you to arrange the table as per the user's choice.

```python
df_20 = pd.DataFrame({'S.No': [1,2,3,4,5], 'A':['apple','ant','ankle','air','arm'], 'B':['ball','bat','bear','balloon','boy']})
print(df_20)
```

```python
df_20.pivot(index='A', columns='B', values='B')
```

```python
df_20.pivot(index="S.No", columns="A")["B"]
```

```python
df_20.pivot(index="S.No", columns="A", values=["A","B"])
```
